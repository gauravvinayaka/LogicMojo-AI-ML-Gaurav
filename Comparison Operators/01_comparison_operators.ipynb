## What comparison operators do
Comparison operators compare two values and return a Boolean: `True` or `False`.


## Equal and not equal


```python
print(2 == 2)
print(2 == 1)
print("apple" == "apple")
print("apple" == "Apple")
```

```python
print(3 != 3)
print(3 != 4)
print("online" != "offline")
```

## Greater than and less than


```python
print(10 > 5)
print(10 > 15)
print(4 < 9)
print(4 < 2)
```

## Greater than or equal to / less than or equal to


```python
print(100 >= 100)
print(100 >= 80)
print(50 <= 75)
print(50 <= 25)
```

## Real-world example: checking a budget


```python
budget = 120
cart_total = 98

print("Cart is within budget:", cart_total <= budget)
print("Cart is over budget:", cart_total > budget)
```

## Real-world example: checking age requirement


```python
age = 19
minimum_age = 18

print("Allowed:", age >= minimum_age)
```

## Comparing strings
Strings are case-sensitive. That means lowercase and uppercase letters are treated differently.


```python
username = "Gaurav"
typed_username = "gaurav"

print(username == typed_username)
print(username.lower() == typed_username.lower())
```

## Comparing variable values


```python
current_temperature = 72
target_temperature = 70

print(current_temperature == target_temperature)
print(current_temperature > target_temperature)
print(current_temperature < target_temperature)
```

## Quick practice


```python
sale_price = 45
original_price = 60

print("Discount applied:", sale_price < original_price)
print("Same price:", sale_price == original_price)
```
