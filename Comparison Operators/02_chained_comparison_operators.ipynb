## Combining comparisons
Python lets you combine multiple comparisons using `and`, `or`, and chained syntax.


## Using `and`
`and` returns `True` only when both conditions are true.


```python
price = 35

print(price > 20 and price < 50)
print(price > 20 and price > 50)
```

## Using `or`
`or` returns `True` when at least one condition is true.


```python
weather = "rainy"

print(weather == "sunny" or weather == "rainy")
print(weather == "snowy" or weather == "windy")
```

## Using `not`
`not` reverses a Boolean value.


```python
store_open = False

print(store_open)
print(not store_open)
```

## Chained comparisons
Instead of writing `x > 1 and x < 10`, Python can write `1 < x < 10`.


```python
x = 5

print(1 < x < 10)
print(1 < x > 10)
```

## Real-world example: valid rating range


```python
rating = 4.5

print("Valid rating:", 0 <= rating <= 5)
```

## Real-world example: shipping discount


```python
cart_total = 75
member = True

free_shipping = cart_total >= 50 and member
print("Free shipping:", free_shipping)
```

## Real-world example: weekend check


```python
day = "Saturday"

is_weekend = day == "Saturday" or day == "Sunday"
print("Weekend:", is_weekend)
```

## Combining several conditions


```python
battery_percent = 82
plugged_in = False

ready_for_trip = battery_percent >= 80 and not plugged_in
print("Ready for trip:", ready_for_trip)
```

## Quick practice


```python
score = 88

print("Passing:", score >= 70)
print("B grade range:", 80 <= score < 90)
print("Needs review:", score < 70 or score > 100)
```
