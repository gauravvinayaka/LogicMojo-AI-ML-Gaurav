A concise practice notebook with real-world examples and visible printed answers.


## Creating sets
Sets only keep unique values.


```python
items = {"apple", "banana", "apple", "orange"}
print(items)
```

## Adding and removing


```python
items = {"apple", "banana", "orange"}
items.add("grapes")
items.remove("banana")

print(items)
```

## Set operations


```python
customers_monday = {"Ava", "Ben", "Carlos"}
customers_tuesday = {"Ben", "Dana", "Eli"}

print(customers_monday.union(customers_tuesday))
print(customers_monday.intersection(customers_tuesday))
print(customers_monday.difference(customers_tuesday))
```

## Booleans


```python
print(True)
print(False)
```

```python
is_open = True
has_coupon = False

print(is_open and has_coupon)
print(is_open or has_coupon)
print(not has_coupon)
```

## Comparisons return booleans


```python
print(10 > 5)
print(10 == 5)
print("apple" == "apple")
```
