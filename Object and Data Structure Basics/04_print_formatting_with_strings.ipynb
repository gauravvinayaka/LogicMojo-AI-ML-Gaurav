# 04 - Print Formatting with Strings

A concise practice notebook with real-world examples and visible printed answers.


## Using variables inside printed messages


```python
name = "Sam"
age = 29

print("My name is {} and I am {} years old.".format(name, age))
```

## Reordering with `.format()`


```python
print("The {2} {1} {0}".format("fox", "brown", "quick"))
```

## f-strings


```python
product = "headphones"
price = 59.99

print(f"The {product} cost ${price}.")
```

## Number formatting


```python
total = 127.4567
print(f"Total: ${total:.2f}")
```

```python
discount = 0.20
print(f"Discount: {discount:.0%}")
```

## Real-world receipt example


```python
item = "coffee maker"
quantity = 2
unit_price = 34.95
subtotal = quantity * unit_price

print(f"Item: {item.title()}")
print(f"Quantity: {quantity}")
print(f"Subtotal: ${subtotal:.2f}")
```
