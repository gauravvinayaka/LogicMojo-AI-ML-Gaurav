## Creating dictionaries
Dictionaries store key-value pairs.


```python
car = {"brand": "Toyota", "model": "Camry", "year": 2022}
print(car)
```

## Accessing values by key


```python
car = {"brand": "Toyota", "model": "Camry", "year": 2022}
print(car["brand"])
print(car["year"])
```

## Updating and adding values


```python
car = {"brand": "Toyota", "model": "Camry", "year": 2022}
car["year"] = 2023
car["color"] = "blue"

print(car)
```

## Dictionary methods


```python
profile = {"name": "Jordan", "city": "Austin", "active": True}

print(profile.keys())
print(profile.values())
print(profile.items())
```

## Nested dictionaries


```python
inventory = {
    "laptop": {"price": 899, "stock": 12},
    "tablet": {"price": 399, "stock": 20}
}

print(inventory["laptop"]["price"])
print(inventory["tablet"]["stock"])
```
