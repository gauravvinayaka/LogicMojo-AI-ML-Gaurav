## Writing to a file
Use `with open(...)` so Python closes the file automatically.


```python
with open("daily_notes.txt", "w") as file:
    file.write("Buy groceries\n")
    file.write("Call the bank\n")

print("File created.")
```

## Reading the whole file


```python
with open("daily_notes.txt", "r") as file:
    contents = file.read()

print(contents)
```

## Reading lines


```python
with open("daily_notes.txt", "r") as file:
    lines = file.readlines()

print(lines)
```

## Appending to a file


```python
with open("daily_notes.txt", "a") as file:
    file.write("Schedule car service\n")

with open("daily_notes.txt", "r") as file:
    print(file.read())
```

## Real-world example: simple checklist file


```python
checklist = ["Pack laptop", "Charge phone", "Print ticket"]

with open("travel_checklist.txt", "w") as file:
    for task in checklist:
        file.write(task + "\n")

with open("travel_checklist.txt", "r") as file:
    print(file.read())
```
