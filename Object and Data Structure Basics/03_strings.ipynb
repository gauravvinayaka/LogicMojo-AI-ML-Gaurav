## Length


```python
city = "Chicago"
print(len(city))
```

## Indexing and slicing


```python
word = "restaurant"
print(word[0])
print(word[3])
print(word[-1])
```

```python
word = "restaurant"
print(word[0:4])
print(word[4:])
print(word[:6])
```

## String methods


```python
name = "alex johnson"
print(name.upper())
print(name.title())
```

```python
message = "   order confirmed   "
print(message.strip())
```

## Real-world example: email username


```python
email = "maya.parker@example.com"
username = email.split("@")[0]

print(username)
```
