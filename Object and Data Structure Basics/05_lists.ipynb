## Creating lists
Lists store items in order.


```python
shopping_list = ["apples", "bread", "milk"]
print(shopping_list)
```

## Indexing and slicing


```python
shopping_list = ["apples", "bread", "milk", "eggs"]
print(shopping_list[0])
print(shopping_list[-1])
print(shopping_list[1:3])
```

## Changing list items


```python
shopping_list = ["apples", "bread", "milk"]
shopping_list[1] = "bagels"

print(shopping_list)
```

## Adding and removing items


```python
tasks = ["laundry", "groceries"]
tasks.append("clean kitchen")
print(tasks)

removed_task = tasks.pop()
print("Removed:", removed_task)
print("Remaining:", tasks)
```

## Sorting


```python
scores = [88, 92, 75, 100, 84]
scores.sort()

print(scores)
```

## Nested lists


```python
matrix = [[1, 2, 3], [4, 5, 6]]
print(matrix[0])
print(matrix[1][2])
```
