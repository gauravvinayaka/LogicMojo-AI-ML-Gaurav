## Real-world example: splitting a dinner bill


```python
meal_total = 86.40
people = 4
tip_rate = 0.18

tip = meal_total * tip_rate
grand_total = meal_total + tip
amount_each = grand_total / people

print("Tip:", round(tip, 2))
print("Grand total:", round(grand_total, 2))
print("Each person pays:", round(amount_each, 2))
```
