## Creating tuples
Tuples are ordered collections that cannot be changed.


```python
location = (41.8781, -87.6298)
print(location)
```

## Indexing tuples


```python
location = (41.8781, -87.6298)
print(location[0])
print(location[1])
```

## Tuple methods


```python
colors = ("red", "blue", "red", "green")
print(colors.count("red"))
print(colors.index("green"))
```

## Tuple unpacking


```python
dimensions = (1920, 1080)
width, height = dimensions

print("Width:", width)
print("Height:", height)
```

## Why use tuples?
Use a tuple when grouped data should stay fixed, such as GPS coordinates or screen size.


```python
screen_size = (13.3, "inches")
print(screen_size)
```
